# **RDD**: Resilient Distributed Datasets in Python

## 1. Introduction

This notebook shows how to connect Jupyter notebooks to a Spark cluster to process data using the Spark Python API with RDD.

## 2. Context

```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf() \
        .setAppName("RDD Example") \
        .setMaster("spark://spark-master:7077") \
        .set("spark.executor.memory", "512m")
```

```python
spark = SparkContext(conf=conf)
```

## 3. Actions

```python
lines = spark.textFile("data/logs.txt")
```

```python
words = lines.flatMap(lambda line: line.split())
```

```python
filtered = words.filter(lambda word: len(word) > 3)
```

```python
pairs = filtered.map(lambda word: (word, 1))
```

## 4. Transformations

```python
counts = pairs.reduceByKey(lambda a, b: a + b)
```

```python
result = counts.collect()
```

```python
print(result)
```
